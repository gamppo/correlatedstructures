```python
#import libraries
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# import the frustration data
datapath = "/usr/users/olivia/data/PDB/DATA_Olivia/dssp_csv/"


for file in os.listdir(datapath):

    protein=file[0:4]
    #import frustration data
    frustData =pd.read_csv(f'/usr/users/olivia/data/PDB/DATA_Olivia/frust/{protein}.done/FrustrationData/{protein}.pdb_configurational_5adens', sep = ' ')

    #import the dssp data
    dsspData = pd.read_csv(f'/usr/users/olivia/data/PDB/DATA_Olivia/dssp_csv/{protein}_dssp.csv')

    #import the angles and the ang/ dist correlations
    corrData = pd.read_csv(f'/usr/users/olivia/data/PDB/DATA_Olivia/PhiPsiAngDistCor/{protein}_Correlations.csv', index_col=0)
    #make a protein overview  file that has all the correlations, frust data and dssp data

    #get the number of conformers so that the correct number of reps can be added for the frust and dssp data
    nConformers= int(corrData.iat[-1,0]+1)
    print(nConformers)
    print(corrData)
    #make the dssp data the correct length
    dssp=pd.concat([dsspData]*nConformers, ignore_index=True)

    #make the frust data the correct length
    frust= pd.concat([frustData]*nConformers, ignore_index=True)
    #make the overview dataframe that contains the columns from all 3 dataframes
    overview = pd.concat([corrData, dssp.iloc[:,0:2], frust.iloc[:,-3:]], axis=1)
    #save overview to a csv file
    #columns are as followed: conformer_id, residue_id, phi, psi, ang_cor, dist_cor, AA, ss, high frustration, neutral frustration, minimal frustration
    overview.to_csv(f'/usr/users/olivia/data/PDB/DATA_Olivia/Overview/{protein}_overview.csv', index= False)


```
